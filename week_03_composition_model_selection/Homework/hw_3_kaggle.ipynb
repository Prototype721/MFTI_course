# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
# "Churn" соотношение с cat_cols

fig, axes = plt.subplots(4, 4, figsize=(30, 30))
axes = axes.flatten()

for ind, cat_col in enumerate(cat_cols):
  CrosstabResult = pd.crosstab(index=data[cat_col], columns=data["Churn"])
  CrosstabResult.plot.bar(ax=axes[ind], title=cat_col, legend=True)
  axes[ind].set_ylabel("Count")

plt.show()
```

```python
# "Churn" соотношение с num_cols

fig, axes = plt.subplots(1, 3, figsize=(17, 8))
axes = axes.flatten()

for ind, num_col in enumerate(num_cols):
  work_data = data.loc[:, [num_col, "Churn"]]
  work_data_0 = work_data[work_data["Churn"] == 0][num_col]
  work_data_1 = work_data[work_data["Churn"] == 1][num_col]
  axes[ind].hist(work_data_0, alpha=0.25, color="blue", label="Churn: 0")
  axes[ind].hist(work_data_1, alpha=0.35, color="red", label="Churn: 1")

  mean_0 = np.mean(work_data_0)
  mean_1 = np.mean(work_data_1)

  axes[ind].axvline(mean_0, color='blue', linestyle='dashed', linewidth=2, label='Mean Churn = 0')
  axes[ind].axvline(mean_1, color='red', linestyle='dashed', linewidth=2, label='Mean Churn = 1')

  axes[ind].set_title(num_col)
  axes[ind].set_ylabel("count")
  axes[ind].legend()

plt.show()
```
